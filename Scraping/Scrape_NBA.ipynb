```python
from selenium import webdriver
from bs4 import BeautifulSoup
```

```python
class Player():
    """docstring for Player"""
    def __init__(self):
        self.name = ""
        self.link = ""
        self.Height = ""
        self.Weight = ""
        self.Born = ""
```

```python
def get_player_list():
    driver = webdriver.PhantomJS(executable_path=r'C:/Users/johna/Downloads/phantomjs')
    url = 'http://www.nba.com/players/'
    driver.get(url)
    
    soup = BeautifulSoup(driver.page_source, 'lxml')
    div = soup.find('div', {'id': 'player-left-block'})
    
    player_list = []
    for a in div.find_all('a'):
        new_player = Player()
        new_player.name = a.text.split('#')[0]
        new_player.link = 'http://www.nba.com' + a['href']
        player_list.append(new_player)
    
    for one_player in player_list:
        print(one_player.name)
        print(one_player.link)
        
    driver.quit()
    
    return player_list
```
