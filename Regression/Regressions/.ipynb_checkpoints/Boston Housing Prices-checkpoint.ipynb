# Robust Regression

Linear Regression is heavily impacted by the presence of outliers. 

```python
X = df['RM'].values.reshape(-1, 1)
y = df['MEDV'].values
```

```python
# automatically samples and tries to identify outliers, proceeds to exclude those outliers
# need to define residual threshold manually
from sklearn.linear_model import RANSACRegressor
```

```python
ransac = RANSACRegressor()
```

```python
ransac.fit(X, y)
```

```python
inlier_mask = ransac.inlier_mask_
outlier_mask = np.logical_not(inlier_mask)
```

```python
line_X = np.arange(3, 10, 1)
line_y_ransac = ransac.predict(line_X.reshape(-1, 1))
```

```python
plt.figure(figsize=(12, 10))
plt.scatter(X[inlier_mask], y[inlier_mask], c='blue', marker='o', label='Inliers')
plt.scatter(X[outlier_mask], y[outlier_mask], c='brown', marker='s', label='Outliers')
plt.plot(line_X, line_y_ransac, color='red')
plt.xlabel('Avg # of Rooms per Dwelling')
plt.ylabel("Median Value of owner-occupied homes in $1000s")
plt.legend(fontsize=14)
```

```python
ransac.estimator_.coef_
```

```python
ransac.estimator_.intercept_
```

```python
X = df['LSTAT'].values.reshape(-1, 1)
```

```python
ransac.fit(X, y)
```

```python
inlier_mask = ransac.inlier_mask_
outlier_mask = np.logical_not(inlier_mask)

line_X = np.arange(0, 40, 1)
line_y_ransac = ransac.predict(line_X.reshape(-1, 1))
```

```python
plt.figure(figsize=(12, 10))
plt.scatter(X[inlier_mask], y[inlier_mask], c='blue', marker='o', label='Inliers')
plt.scatter(X[outlier_mask], y[outlier_mask], c='brown', marker='s', label='Outliers')
plt.plot(line_X, line_y_ransac, color='red')
plt.xlabel('% lower status of the population')
plt.ylabel("Median Value of owner-occupied homes in $1000s")
plt.legend(fontsize=14)
```

# Performance Evaluation

```python
from sklearn.model_selection import train_test_split
```

```python
X = df['LSTAT'].values.reshape(-1, 1)
y = df['MEDV'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
y_train_pred = lr.predict(X_train)
```

```python
y_test_pred = lr.predict(X_test)
```

## Residual Analysis

```python
plt.figure(figsize=(12,8))

plt.scatter(y_train_pred, y_train_pred - y_train, c='blue', marker='o', label="Training Data")
plt.scatter(y_test_pred, y_test_pred - y_test, c='red', marker='s', label="Test Data")
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.legend(loc="upper left", fontsize=14)
plt.hlines(y=0,xmin=-10, xmax=50)
plt.xlim([-10, 50])
```

## Mean Squared Error

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_train, y_train_pred)
```

```python
mean_squared_error(y_test, y_test_pred)
```

## Coefficient of Determination, R^2

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_train, y_train_pred)
```

```python
r2_score(y_test, y_test_pred)
```

# All Variables

```python
X = df.iloc[:, :-1].values
y = df['MEDV'].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

lr = LinearRegression()

lr.fit(X_train, y_train)
```

```python
y_train_pred = lr.predict(X_train)
y_test_pred = lr.predict(X_test)
```

```python
plt.figure(figsize=(12,8))

plt.scatter(y_train_pred, y_train_pred - y_train, c='blue', marker='o', label="Training Data")
plt.scatter(y_test_pred, y_test_pred - y_test, c='red', marker='s', label="Test Data")
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.legend(loc="upper left", fontsize=14)
plt.hlines(y=0,xmin=-10, xmax=50)
plt.xlim([-10, 50]);
```

```python
mean_squared_error(y_train, y_train_pred)
```

```python
mean_squared_error(y_test, y_test_pred)
```

```python
r2_score(y_train, y_train_pred)
```

```python
r2_score(y_test, y_test_pred)
```

Using all of the variables to determine our regression line has increased the predictive power of the model and reduced error.
