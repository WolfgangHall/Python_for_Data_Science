The goal of this project is to familiarize myself with Recurrent Neural Networks. The data is a 5-year window of Google stock prices. 

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
training_set = pd.read_csv('Google_Stock_Price_Train.csv')
```

```python
# extracts Open Price
# returns a matrix, 2d numpy array
training_set = training_set.iloc[:,1:2].values
```

```python
# scale features
from sklearn.preprocessing import MinMaxScaler
```

```python
sc = MinMaxScaler()
training_set_scaled = sc.fit_transform(training_set)
```

```python
training_set
```

```python
# Creating a data structure with 20 timesteps and t+1 output
X_train = []
y_train = []
for i in range(20, 1258):
    X_train.append(training_set_scaled[i-20:i, 0])
    y_train.append(training_set_scaled[i, 0])
X_train, y_train = np.array(X_train), np.array(y_train)
```

```python
# reshape inputs
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
```

```python
# Initialize the RNN
from keras.modelsels import Sequential
from keras.layers import Dense
from keras.layers import LSTM
```

```python
regressor = Sequential()
```

```python
# input and LSTM layer
regressor.add(LSTM(units=3, input_shape=(None, 1)))
```

```python
# output layer
regressor.add(Dense(units=1))
```

```python
regressor.compile(optimizer='rmsprop', loss='mean_squared_error')
```
