```python
# percent change
# shift forward 1 so that you can grab previous days price
# tesla['returns'] = (tesla['Close'] / tesla['Close'].shift(1)) - 1
```

```python
# built in pandas functions for above calculation
tesla['returns'] = tesla['Close'].pct_change(1)
ford['returns'] = ford['Close'].pct_change(1)
gm['returns'] = gm['Close'].pct_change(1)
```

```python
tesla['returns'].hist(bins=100, label='Tesla', figsize=(10,8), alpha=0.5)
ford['returns'].hist(bins=100, label='Ford', alpha=0.5)
gm['returns'].hist(bins=100, label='GM', alpha=0.5)
plt.legend()
```

GM and Ford tend to be more stable. Tesla has quite a bit more volatility on a day to day basis.

```python
tesla['returns'].plot(kind='kde', label='Tesla', figsize=(10,8))
ford['returns'].plot(kind='kde', label='Ford', figsize=(10,8))
gm['returns'].plot(kind='kde', label='GM', figsize=(10,8))
plt.legend()
```

## Cumulative Returns

Cumulative return: the aggregate amount an investment has gained or lost over time, independent of the period of time involved.

Takes daily returns into account.

Daily Return: the profit/loss made by the stock compared to the previous day.
    
Value above 1 indicated profit, value below one indicates loss.

Cumulative return does not give the investor insight into the gains they have made till date, especially if the stock is volatile.
    -> computed relative to the day the investment is made

```python
tesla['Cumulative Return'] = (1 + tesla['returns']).cumprod()
ford['Cumulative Return'] = (1 + ford['returns']).cumprod()
gm['Cumulative Return'] = (1 + gm['returns']).cumprod()
```

```python
tesla['Cumulative Return'].plot(label='Tesla', figsize=(16,8))
ford['Cumulative Return'].plot(label='Ford', figsize=(16,8))
gm['Cumulative Return'].plot(label='GM')
plt.legend()
```
